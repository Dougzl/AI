```python
import matplotlib.pyplot as plt
import numpy as np
import math

%matplotlib inline
```

```python
x = np.linspace(-3, 3, 30)
y = x**2
```

```python
plt.plot(x,y, ":")
plt.show()
```

### 颜色： b,g,r,c,m,y,k,w
### 符号：>,<,^,v,X,D,d,1,2,3,4,H,h,p,:,_,-,-.,.,o

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x), 'r--')
plt.plot(x, -np.sin(x), 'g-.')
plt.show()
```

```python
plt.plot(x,y)
plt.title('Sin,Wave')
plt.xlabel('angle', color='r', fontsize=20)
plt.ylabel('sin', color='b', rotation=-60, fontsize=15)
plt.show()
```

```python
# 使用面向对象的方法画图
fig = plt.figure()
# figure类，类似于一个容器，就像画图需要一张画纸
```

```python
ax = fig.add_axes([0,0,1,1])
print(ax)
```

```python
x=np.arange(0, math.pi*2, 0.05)
y=np.sin(x)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_title('sin wave')
ax.set_xlabel('angle')
ax.set_ylabel('sin')
ax.plot(x,y)
plt.show()
```

```python
y = range(1, 8)
x1 = [1 ,15, 25, 30, 59, 90, 100]
x2 = [1, 5, 8, 13, 17, 22, 34]

fig = plt.figure(figsize=(10, 5))
ax = fig.add_axes([0,0,1,1])
ax.plot(x1, y, 'bo-')
ax.plot(x2, y, 'rs--')
ax.legend(labels=('TV', 'Phone'))
plt.show()
```

## 在画布上创建多个子图

```python
ax1 = plt.subplot(212)
ax1.plot(range(12))
ax2 = plt.subplot(211, facecolor = 'y')
ax2.plot(range(12))
plt.show()
```

```python
fig.axlist = plt.subplots(2,2)
```

```python
fig,axlist = plt.subplots(2,2)
x = np.arange(1,5)
axlist[0,0].plot(x, x*x)
axlist[0,1].plot(x, np.exp(x))
axlist[1,0].plot(x, np.sqrt(x))
axlist[1,1].plot(x, np.log10(x))
plt.show()
```

```python
ax1 = plt.subplot2grid((3,3), (0,0), colspan=2, rowspan=1)
ax2 = plt.subplot2grid((3,3), (0,2), rowspan=3)
ax3 = plt.subplot2grid((3,3), (1,0), colspan=2, rowspan=2)

x = np.arange(1, 10)
ax1.plot(x, np.exp(x))
ax1.set_title('square root')

ax3.plot(x, np.log(x))
ax3.set_title('log')

ax2.plot(x, x**2)
ax2.set_title('square')
```

```python
fig = plt.figure()
x = np.arange(0, math.pi * 2, 0.05)

ax1 = fig.add_axes([0,0,1,1])
ax1.plot(x, np.sin(x))
ax1.set_title('sin')

ax2 = fig.add_axes([0.55, 0.65, 0.3, 0.3], facecolor='g')
ax2.plot(x, np.cos(x), 'r')
ax2.set_title('cos')

plt.show()
```

```python
fig, axes = plt.subplots(3,1, figsize = (12,6))

x = np.arange(1, 11)
axes[0].plot(x, x**2)
axes[0].grid(True)

axes[1].plot(x, 1/x)
axes[1].grid(color= 'r', ls= '--', lw=0.3)

axes[2].plot(x, np.exp(x))
axes[2].grid(color = 'y')

axes[2].set_xlabel('x axis')
axes[2].set_ylabel('y axis')

plt.show
```

```python
fig, axes = plt.subplots(1,2, figsize=(12,4))

x = np.linspace(-1, 1, 512)
f = lambda x: (1 - x**2)**0.5
axes[0].plot(x, f(x))
axes[0].plot(x, -f(x))

axes[1].plot(x, f(x))
axes[1].plot(x, -f(x))

axes[1].axis('equal') # 让x轴和y轴相等
axes[1].axis('off')   # 关闭轴线

plt.show()
```
