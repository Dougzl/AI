## 使用matplotlib 画图, 包括直方图, 条形图, 饼图, 散点图, 箱形图, 轮廓图

```python
import matplotlib.pyplot as plt
import numpy as np
import math

import pandas as pd
from IPython.core.pylabtools import figsize
%matplotlib inline
# 上一行代码表示在notebook中显示
```

```python
x = np.arange(0, math.pi * 2, 0.05)
y = np.sin(x)
plt.plot(x, y, 'y', '-.')
plt.show()
```

### 使用面向对象的方法画图
#### figure类, 类似一个容器, 就像画图需要有一张画布一样
#### figure对象通过调用add_axes()方法将Axes对象添加到图中
#### 它返回轴对象并在rect[left, bottom, width, height] 添加一个轴

```python
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
ax.set_title('sin wave')
ax.set_xlabel('angle')
ax.set_ylabel('angle')

ax
```

```python
fig, axes = plt.subplots(1, 2, figsize = (12, 4))
x = np.arange(1, 6)
# 原图
axes[0].plot(x, np.exp(x))
axes[0].plot(x, x**2)

axes[1].plot(x, np.exp(x))
axes[1].plot(x, x**2)
# 对右边的图(axes[1])做处理
axes[1].spines['bottom'].set_color('red')
axes[1].spines['left'].set_color('yellow')
axes[1].spines['right'].set_color(None)
axes[1].spines['top'].set_color(None)

plt.show()
```

```python
# 设置轴线的类型
fig, axes = plt.subplots(1, 2, figsize = (12,4))

x = np.linspace(-1, 1, 200)
f = lambda x:(1-x**2)**0.5
axes[0].plot(x, f(x))
axes[0].plot(x, -f(x))
# 不圆是 x轴和y轴不完全相等
axes[1].plot(x, f(x))
axes[1].plot(x, -f(x))

axes[1].axis('equal')
axes[1].axis('off')
plt.show()
```

### 箱形图
- 箱形图也成为 须状图, 显示包括5个内容, [最小值, 第一个四分位数, 中位数, 第三个四分位数, 最大值]的一个数据的摘要
- 在方块图中绘制从第一个四方位数到第三个四方位数的方框, 垂直线穿过中间的框

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_axes([0,0,1,1])

np.random.seed(10)
col_1 = np.random.normal(100, 10, 200)# 高斯分布, 括号中的三个数表示均值, 标准差, 200个数字
col_2 = np.random.normal(80, 20, 200)# 高斯分布, 括号中的三个数表示均值, 标准差, 200个数字
col_3 = np.random.normal(90, 15, 200)# 高斯分布, 括号中的三个数表示均值, 标准差, 200个数字
col_4 = np.random.normal(150, 25, 200)# 高斯分布, 括号中的三个数表示均值, 标准差, 200个数字

ax.boxplot([col_1, col_2, col_3, col_4])

plt.show()
```

### 轮廓图
- 是一种在二维平面上显示三维表面的方法
- 它绘制了Y轴上两个预测变量 X Y 和轮廓的相应变量Z
- 如果要查看Z 如何随两个输入变量 X和Y的变化,则轮廓图就非常合适, 比如 z=f(x,y)

```python
fig = plt.figure(figsize= (10, 5))
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')

x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)

x,y = np.meshgrid(x,y)
z = np.sqrt(x**2 + y**2)

ct = ax.contourf(x,y,z)
fig.colorbar(ct)

plt.show()
```

### 三维曲面图
- 三维曲面图展示了三个变量之间的关系, 以及其在三维空间中的分布形态
- 在三维曲面图中, 数据点被映射到一个三维坐标系中的曲面上, 通过曲面的形状, 高度或者颜色来表示数据特征和变化趋势

```python
import plotly.graph_objects as go

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x,y = np.meshgrid(x,y)
z = np.sin(np.sqrt(x**2+y**2))

fig = go.Figure(data= [go.Surface(z= z, y= y, x= x)])
fig.update_layout(title= '3D Surface Plot', scene= dict(xaxis_title= 'X-axis', yaxis_title='Y-axis', zaxis_title='Z-axis'))

fig.show()
```
